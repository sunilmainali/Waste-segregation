```python
from ultralytics import YOLO
model = YOLO("yolov9c.pt")
```

```python
model.train(
    data="data.yaml",
    epochs=100,
    imgsz=640,
    project="runs/train",
    name="Optimized_Model",
    exist_ok=True,
    device=0
)
```

```python
model.val(data="data.yaml", split="train")
```

```python
# Export model to ONNX format
model.export(format='onnx')
```
